```python
def INFO(df):
    info = []

    for col in df.columns:
        count = len(df[col])
        NAN_values = df[col].isna().sum()
        percent = (NAN_values / count) * 100
        data_type = df[col].dtype
        if data_type not in [int,float]:
            column_type = "Categorical"
            Max = "Not Applicable"
            Min = "Not Applicable"
        else:
            column_type = "Numirical"
            Max = max(df[col])
            Min = min(df[col])
        try:
            n_uniques =df[col].nunique()
            ratio = count// n_uniques
        except:
            n_uniques = "Not Applicable"
            ratio = "Not Applicable"
        info.append([col,data_type,column_type, count, NAN_values, percent,n_uniques,ratio,Max, Min])

    col_info_df = pd.DataFrame(info, columns=['Column','Data Type','Column Type', 'Total', 'Missing', 'Percent Missing','Number of Uniques','Ratio of uniqus','Max','Min'])

    return col_info_df
```

```python
def impute_data(df):
    # We will use a simple equation to impute time "total_time = Cook_time + Prep_time" and work around this 
    # Fill in missing total_time values
    df.loc[df["total_time"].isna(), "total_time"] = df["prep_time"] + df["cook_time"]
    
    # Fill in missing prep_time values
    df.loc[df["prep_time"].isna(), "prep_time"] = df["total_time"] - df["cook_time"]
    
    # Fill in missing cook_time values
    df.loc[df["cook_time"].isna(), "cook_time"] = df["total_time"] - df["prep_time"]
    
    # IF there are any discrepancy in total_time we are going to use the same equation to rectifie it
    # Adjust cook_time if total_time is greater than prep_time + cook_time
    df.loc[df["total_time"] > (df["prep_time"] + df["cook_time"]), "cook_time"] = df["total_time"] - df["prep_time"]
    
    # The ratings wtih Null values will be imputed to 0.0 denoting that no one has yet tried out this recipe 
    df.rating = df.rating.fillna(0.0)
    
    # At the end we will be imputing Null values in prep and cook time after the above imputations with 0.0
    df.prep_time = df.prep_time.fillna(0.0)
    df.cook_time = df.cook_time.fillna(0.0)
    
    # reseting the index of the dataframe after all the imputations
    df = df.reset_index(drop = True)
    
    # Changing the names of continents
    try:
        df.Contient = df.Contient.replace({"North_America":"North America","Canada":"North America",'asia':'Asia','australia':'Australia',"europe":"Europe"})
    except:
        pass
    # Returning the cleaned and imputed dataframe
    return df
```

```python
# nutrients string to dict
def nutrients_dict(nutrients):
    nutrients = ast.literal_eval(nutrients)
    return nutrients
```

```python
# replace fraction to decimal
def replace_fractions(ingredients):
#     ingredients = ast.literal_eval(ingredients)
    # define a dictionary to map all fraction symbols to their decimal values
    fraction_dict = {
        '\u00BC': '0.25', 
        '\u00BD': '0.5', 
        '\u00BE': '0.75', 
        '\u2150': '0.1', 
        '\u2151': '0.2', 
        '\u2152': '0.3', 
        '\u2153': '0.333', 
        '\u2154': '0.4', 
        '\u2155': '0.167', 
        '\u2156': '0.833', 
        '\u2157': '0.375', 
        '\u2158': '0.625', 
        '\u2159': '0.111', 
        '\u215A': '0.222', 
        '\u215B': '0.125', 
        '\u215C': '0.167', 
        '\u215D': '0.2', 
        '\u215E': '0.25'
    }
    
    # create a new list to store the updated ingredients
    updated_ingredients = []
    
    # use regular expressions to search for all fraction symbols in each ingredient and replace them with their decimal values
    for ingredient in ingredients:
        for symbol, value in fraction_dict.items():
            pattern = re.compile(symbol)
            ingredient = re.sub(pattern, value, ingredient)
        updated_ingredients.append(ingredient)
    return updated_ingredients
```

```python
# removing units from columns
def extract_digit(data,column_list):
    try:
        for col in column_list:
            if col in data.columns:
                data[col] = data[col].str.extract('(\d+\.?\d*)').astype(float)
                print(f'{col}.....done!')
            else:
                print(f'{col}.....not found!')
    except:
        print("All ready Extracted")
```

```python
import nltk
from nltk import word_tokenize, pos_tag

def extract_nouns(ingredients):
#     ingredients = ' '.join(ingredients)
    all_nouns = set()

    def extract_nouns_helper(sentence, existing_nouns):
        nouns = set()
        for word, tag in pos_tag(word_tokenize(sentence)):
            if tag.startswith('N'):
                if word not in existing_nouns:
                    nouns.add(word.lower())
        return nouns

    for string in ingredients:
        all_nouns.update(extract_nouns_helper(string, all_nouns))

    return list(set(all_nouns)-set(measurement_units))
```
